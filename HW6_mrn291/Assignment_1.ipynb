```python
#make a list of the csv files
file_list = os.listdir(os.getcwd())
file_list = [file for file in file_list if file[-3:]=='csv']
```

```python
#read in csvs
frame_dict = {}
for file in file_list:
    file_name = file[0:2]+"_data"
    frame = pd.read_csv(file)
    frame_dict[file_name] = frame

frame_dict.keys()
```

```python
#show manhattan 10 rows and make a manhattan only dataframe since we're only using MN and BK
MN_data = frame_dict['MN_data']
MN_data.head()
```

```python
#show brooklyn 10 rows and make a brooklyn only dataframe since we're only using MN and BK
BK_data = frame_dict['BK_data']
BK_data.head()
```

```python
#show all manhattan columns
MN_data.columns
```

```python
#show all brooklyn columns
BK_data.columns
```

```python
#drop the columns except for zip and landuse
MN_data.drop(MN_data.columns.difference(['ZipCode','LandUse']), axis=1, inplace=True)
BK_data.drop(BK_data.columns.difference(['ZipCode','LandUse']), axis=1, inplace=True)
```

```python
MN_data.head()
```

```python
BK_data.head()
```

```python
#group the manhattan data by landuse category
mngrouped = MN_data.groupby('LandUse').count().reset_index()
mngrouped
```

```python
#group the bk data by landuse category
bkgrouped = BK_data.groupby('LandUse').count().reset_index()
bkgrouped
```

```python
#change the landuse categories into integers
mngrouped['LandUse'] = mngrouped['LandUse'].astype(int)
bkgrouped['LandUse'] = bkgrouped['LandUse'].astype(int)
bkgrouped.head()
```

```python
#merge the two dataframes on landuse
both = pd.merge(mngrouped, bkgrouped, how='inner', on='LandUse')
both.columns
```

```python
#give better column names
both.rename(columns={'ZipCode_x':'mn', 'ZipCode_y':'bk'}, inplace=True)
```

```python
#set up land use values to descriptive tag
names = ['One & Two Family Buildings','Multi Family Walk Up Buildings', 'Multi Family Elevator Buildings', 'Mixed Residential & Commercial Buildings', 'Commercial & Office Buildings', 'Industrial & Manufacturing', 'Transportation & Utility', 'Public Facilities & Institutions', 'Open Space & Outdoor Recreation', 'Parking Facilities', 'Vacant Land']
data_source = {'LandUse':range(11), 'LandUseName': names}
descriptive = pd.DataFrame(data=data_source)
both = both.merge(descriptive, how='inner', on='LandUse')
both.drop(labels='LandUse', axis=1, inplace=True)
both
```

```python
#calculate the land use ratio
both['landuse_ratio'] = (both['mn'] / both['bk']).astype(float)
both
```

```python
#define a function to determine if the ratio is greater than or less than 1 then apply it to the landuse ratio
def winner(ratio):
    if ratio > 1:
        return True
    else:
        return False

both['Manhattan_Wins'] = both['landuse_ratio'].apply(winner)
```
